# Analysing Using Charts

```python
plt.figure(figsize=(10,5))
sns.set_style('white')
sns.countplot(x='Survived',hue='Sex' ,data=data, palette="autumn")
plt.figure(figsize=(16,9))
```




# One hot Encoding

```python
sex = pd.get_dummies(data['Sex'],drop_first=True)
```

```python
sex.head()
```

```python
merge = pd.concat([data,sex],axis='columns')
```

```python
final = merge.drop(["Sex"],axis='columns')
```

```python
final
```
